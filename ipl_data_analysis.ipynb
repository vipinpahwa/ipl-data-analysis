```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ball_df=pd.read_csv('Ball_by_Ball.csv')
match_df=pd.read_csv('Match.csv')
player_df=pd.read_csv('Player.csv')
player_match_df=pd.read_csv('Player_Match.csv')
season_df=pd.read_csv('Season.csv')
team_df=pd.read_csv('Team.csv')
```

```python
##to display the frequency of IPL matches in particular stadiums
match_df['Venue_Name'].value_counts().plot(kind='bar')
```

```python
match_df=match_df.merge(team_df,how='left',left_on='Match_Winner_Id',right_on='Team_Id')
```

```python
##to display the matches won by a particular team in a particular season.
season=input('enter season no:')
season=int(season)
match_df[match_df['Season_Id']==season]['Team_Name'].value_counts().plot(kind='bar')
```

```python
##to display the number of matches won either by runs or by wickets or drawn in a particular season.
season=input('enter season no:')
season=int(season)
match_df[match_df['Season_Id']==season]['Win_Type'].value_counts().plot(kind='bar')
```

```python
##to display the number of players of each country playing in the IPL.
player_df['Country'].value_counts().plot(kind='bar')
```

```python
player_match_df=player_match_df.merge(player_df,how='left',left_on='Player_Id',right_on='Player_Id')
```

```python
player_match_df=player_match_df.merge(team_df,how='left',left_on='Team_Id',right_on='Team_Id')
```

```python
##to display the playing 11 of both the teams playing a particular match identified by the match id.
mid=input()
mid=int(mid)
player_match_df[player_match_df['Match_Id']==mid][['Team_Name','Player_Name']]
```

```python
##to diplay the man of the match of a particular match.
mid=input()
mid=int(mid)
id=match_df[match_df['Match_Id']==mid]['Man_Of_The_Match_Id']
id=int(id)
player_df[player_df['Player_Id']==id]['Player_Name']
```

```python
##to display the winning teams of the matches played on a particular date
date=input('Enter date:')
l=[]
l=match_df[match_df['Match_Date']==date]['Match_Id']
for i in l:
    print(match_df[match_df['Match_Id']==i]['Team_Name'])
```

```python
##to display the no of times teams elected to either field or bat in a season.
season=input('enter season no:')
season=int(season)
match_df[match_df['Season_Id']==season]['Toss_Decision'].value_counts().plot(kind='bar')
```

```python
##to display the number of times the teams elected to either field or bat at a particular stadium.
venue=input('enter venue:')
match_df[match_df['Venue_Name']==venue]['Toss_Decision'].value_counts().plot(kind='bar')
```
